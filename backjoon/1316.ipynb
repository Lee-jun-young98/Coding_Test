```python
count = 0

def check_al(word, a, b):
    if b > len(word)-1:
        return 1    
    if word[a] == word[b]:
        check_al(word,a,b+1)
        return 1
    else:
        return 0
    

def check(word):
    for j in range(len(word)):
        if word[j] in word[j+1:]:
            word_count = check_al(word, j, j+1)
            return word_count
    return 1
            

for _ in range(int(input())):
    word = list(map(str, input()))
    count += check(word)

print(count)
```
