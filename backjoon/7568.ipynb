```python
n = int(input())
dungcis = [list(map(int, input().split(" "))) for _ in range(n) ]
```

```python
dungcis
```

```python
for i in range(len(dungcis)):
    grade = 1
    for j in range(len(dungcis)):
        if dungcis[i][0] < dungcis[j][0] and dungcis[i][1] < dungcis[j][1]:
            grade = grade + 1
    print(grade, end = " ")

```
