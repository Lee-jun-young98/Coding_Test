```python
n, m = map(int, input().split(" "))
a = []
b = []
for i in range(n):
    a.append(list(map(int, input().split(" "))))

m, k = map(int, input().split(" "))
for i in range(m):
    b.append(list(map(int, input().split(" "))))


def matrix_multiple(a, b, n, k):
    array = [[0]*k for _ in range(n)]
    for array_index in range(n):
        for i in range(n):
            for j in range(len(b[0])):
                array[array_index][i] += a[i][j] * b[j][i]
    
    for i in array:
        print("\t")
        for j in i:
            print(j, end = " ")

            

matrix_multiple(a, b, n, k)
```

```python
def solution(a, b):
    r = len(a)
    c = len(b[0])
    answer = [[0]*c for _ in range(r)]

    for i in range(r):
        for j in range(c):
            for k in range(len(a[0])):
                answer[i][j] += a[i][k]*b[k][j]
   
    for i in answer:
        print("\t")
        for j in i:
            print(j, end = " ")
```

```python
solution(a, b)
```
