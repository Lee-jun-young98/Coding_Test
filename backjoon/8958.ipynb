```python
n = int(input())
array = []
for i in range(n):
    array.append(list(map(str, input())))

for i in array:
    correct = 0
    true_correct = 0
    for j in i:
        if j == 'O':
            true_correct += 1
            correct += true_correct
        else:
            true_correct = 0
    print(correct)
```
