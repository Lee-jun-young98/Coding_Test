```python
n = int(input())

count = 0
correct_number = n

while True:
    if int(n) < 10 and int(n) >=0:
        n = "0" + str(n)
    first_number, second_number = int(n) // 10, int(n) % 10
    middle_number = first_number + second_number
    n = (second_number * 10) + (middle_number % 10)
    count += 1

    if correct_number == n:
        print(count)
        break



```
