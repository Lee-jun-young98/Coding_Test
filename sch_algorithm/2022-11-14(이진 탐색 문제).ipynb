# Q.27 정렬된 배열에서 특정 수의 개수 구하기
- N개의 원소를 포함하고 있는 수열이 오름차순으로 정렬되어 있음
- 이 수열에서 x가 등장하는 횟수를 계산
- ex) {1,1,2,2,2,2,3}이 있을 때 x = 2라면, 현재 수열에서 값이 2인 원소가 4개이므로 4를 출력
- x인 원소가 하나도 없다면 -1을 출력

```python
from bisect import bisect_right, bisect_left

n, x = map(int, input().split(" "))
arr = list(map(int, input().split(" ")))
print(arr)
```

```python
def count_by_arange(arr, x):
    if x in arr:
        count_right = bisect_right(arr, x)
        count_left = bisect_left(arr, x)
        return count_right - count_left
    else:
        return -1
```

```python
count_by_arange(arr, 4)
```

# Q.28 고정점 찾기
- 고정점이란, 수열의 원소 중에서 값과 인덱스가 동일한 원소를 의미함
- ex) a = {-15, -4, 2, 8, 13}이 있을 때 a[2] = 2이므로, 고정점은 2가 된다
- 고정점이 있으면 고정점을 출력하고 없으면 -1을 출력함

```python
n = int(input())
arr = list(map(int, input().split(" ")))
```

```python
def find_fix_point(arr, n):
    for i in range(n):
        if arr[i] == i:
            return arr[i]
    return -1
```

```python
find_fix_point(arr, n)
```

# Q.29 공유기 설치
- 도현이의 집 N개가 수직선 위에 있습니다. 각각의 집의 좌표는 x1, x2, ..., xn이고, 집 여러개가 같은 좌표를 가지는 일은 없습니다
- 공유기 C개 설치하려고 함
- 한 집에는 공유기 하나 설치, 가장 인접한 두 공유기 사이의 거리를 가능한 크게 설치할려 함
- C개의 공유기를 N개의 집에 적당히 설치해서, 가장 인접한 두 공유기 사이의 거리를 최대로 하는 프로그램을 작성하시오

```python
n, c = list(map(int, input().split(" ")))
array = []
for _ in range(n):
    array.append(int(input()))

array.sort() # 이진 탐색 수행을 위해 정렬 수행

start = 1 # 가능한 최소 거리(min gap)
end = array[-1] - array[0] # 가능한 최대 거리(max gap)
result = 0

while (start <= end):
    mid = (start + end) // 2 # mid는 가장 인접한 두 공유기 사이의 거리(gap)를 의미
    value = array[0]
    count = 1
    # 현재의 mid값을 이용해 공유기를 설치
    for i in range(1, n): # 앞에서부터 차근차근 설치
        if array[i] >= value + mid:
            value = array[i]
            count += 1
    if count >= c: # C개 이상의 공유기를 설치할 수 있는 경우, 거리를 증가
        start = mid + 1
        result = mid # 최적의 결과를 저장
    else: # C개 이상의 공유기를 설치할 수 없는 경우, 거리를 감소
        end = mid - 1

print(result)
```

#Q.30 가사 검색
- 노래 가사에 사용된 단어들 중에 특정 키워드가 몇 개 포함되어 있는지 찾으시오
- "?"는 와일드 카드 문자 -> ex)"fro??"는 "frodo","front","frost"가 매칭됨 

```python
from bisect import bisect_left, bisect_right

# 값이 [left_value, right_value]인 데이터의 개수를 반환하는 함수
def count_by_arange(a, left_value, right_value):
    right_index = bisect_right(a, right_value)
    left_index = bisect_left(a, left_value)
    
    return right_index - left_index

# 모든 단어를 길이마다 나누어서 저장하기 위한 리스트
array = [[] for _ in range(10001)]

# 모든 단어를 길이마다 나누어서 뒤집어 저장하기 위한 리스트
reversed_array = [[] for _ in range(10001)]

def solution(words, queries):
    answer = []
    for word in words: # 모든 단어를 접미사 와일드카드 배열, 접두사 와일드카드 배열에 각각 삽입
        array[len(word)].append(word) # 단어를 삽입
        reversed_array[len(word)].append(word[::-1]) # 단어를 뒤집어서 삽입
    
    for i in range(10001): # 이진 탐색을 수행하기 위해 각 단어 리스트 정렬 수행
        array[i].sort()
        reversed_array[i].sort()

    for q in queries: # 쿼리를 하나씩 확인하며 처리
        if q[0] != "?": # 접미사에 와일드카드가 붙은 경우
            res = count_by_arange(array[len(q)], q.replace('?', 'a'), q.replace("?", "z"))
        else: # 접두사에 와일드카드가 붙은 경우
            res = count_by_arange(reversed_array[len(q)], q[::-1].replace('?', 'a'), q[::-1].replace('?', 'z'))
        answer.append(res)
        
    return answer
```

```python
solution(["frodo","front","frost","frozen","frame","kakao"], ["fro??", "????o", "fr???", "fro???", "pro?"])
```
