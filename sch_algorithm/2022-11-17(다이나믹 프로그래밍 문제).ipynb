# 31. 금광
- n x m 크기의 금광이 있을 때 첫번째 열부터 금을 캐기 시작
- 첫번째 열의 어느 행에서부터나 시작 가능
- 오른쪽 위, 오른쪽, 오른쪽 아래 3가지 중 하나의 위치로 이동
- 가장 많은 크기의 금광을 얻어야 함

```python
test_count = int(input())
n, m = map(int, input().split(" "))
test1 = []
test2 = []
arr = list(map(int, input().split(" ")))
```

```python
test1 = []
for i in range(n):
    test1.append(arr[i*4:(i+1)*4])
```

```python
test1
```

```python
for i in range(1, m):
    for j in range(n):
        if j == 0: # 제일위
            test1[j][i] = max(test1[j][i-1], test1[j+1][i-1] + test1[j][i])
        elif j == n-1: # 하단
            test1[j][i] = max(test1[j-1][i-1] + test1[j][i], test1[j][i-1] + test1[j][i])
        else: # 중단
            test1[j][i] = max(test1[j-1][i-1] + test1[j][i], test1[j][i-1] + test1[j][i], test1[j+1][i-1] + test1[j][i])

ans = 0
for i in range(n):
    ans = max(ans, test1[i][m-1]) 
print(ans)       
```

```python
test_count = int(input())

for _ in range(test_count):
    n, m = map(int, input().split(" "))
    arr = list(map(int, input().split(" ")))
    test = []
    for i in range(n):
        test.append(arr[i*4:(i+1)*4])

    for i in range(1, m):
        for j in range(n):
            if j == 0: # 제일위
                test[j][i] = max(test[j][i-1], test[j+1][i-1] + test[j][i])
            elif j == n-1: # 밑
                test[j][i] = max(test[j-1][i-1] + test[j][i], test[j][i-1] + test[j][i])
            else: # 가운데
                test[j][i] = max(test[j-1][i-1] + test[j][i], test[j][i-1] + test[j][i], test[j+1][i-1] + test[j][i])

    ans = 0
    for i in range(n):
        ans = max(ans, test[i][m-1]) 
    print(ans)       

```
