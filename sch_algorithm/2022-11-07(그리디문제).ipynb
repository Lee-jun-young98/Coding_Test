# 1. 모험가 길드
 - N명의 모험가를 대상으로 '공포도'를 측정함
 - 공포도가 X인 모험가는 반드시 X명 이상으로 구성한 모험가 그룹에 참여해야 여행을 떠날 수 있음
 - 여행을 떠날 수 있는 그룹의 최대값을 구하는 프로그램을 만드시오

```python
n = int(input())
guild = sorted(list(map(int, input().split(" "))))
guild
```

```python
result = 0 # 총 그룹의 수
count = 0 # 현재 그룹에 포함된 모험가의 수

for i in guild: # 공포도를 낮은 것부터 하나씩 확인하며
    count += 1
    if count >= i:
        result += 1
        count = 0

print(result)
```

# 2. 곱하기 혹은 더하기
- 각 자리가 숫자(0부터 9)로만 이루어진 문자열 S가 주어졌을 때, 왼쪽부터 오른쪽으로 하나씩 모든 숫자를 확인하며 숫자 사이에 'X' 혹은 '+' 연산자를 넣어 결과적으로 만들어질 수 있는 가장 큰 값을 구하시오

```python
num = list(map(int, input()))
```

```python
num
```

```python
total_sum = 0

for i in num:
    if i <= 1 or total_sum == 0:
        total_sum += i
    else:
        total_sum *= i

print(total_sum)
```

# 3. 문자열 뒤집기
- 다솜이는 0과 1로만 이루어진 문자열 S를 가지고 있습니다. 이 문자열 S에 있는 모든 숫자를 전부 같게 만들려고 함
- 문자열 S가 주어졌을 때, 다솜이가 해야하는 횟수의 최소 횟수를 출력

```python
string = list(map(str, input()))
string
```

```python
count0 = 0
count1 = 0
for i in range(len(string)-1):
    if string[i] != string[i+1]:
        if string[i+1] == '1':
            count0 += 1
        else:
            count1 += 1

print(min(count0, count1))
```

# 5. 볼링공 고르기

```python
n, m = map(int, input().split(" "))
arr = list(map(int, input().split(" ")))
arr
```

```python
arr.sort()
```

```python
arr
```

```python
count = 0
for i in range(n-1):
    for j in range(i, n):
        if arr[i] != arr[j]:
            count += 1

print(count)

```

# 6. 무지의 먹방 라이브
- 회전판에 먹어야할 N개의 음식이 있음
- 각 음식에서는 1부터 N까지의 번호가 붙어있으며, 각 음식을 먹을 때는 일정 시간이 소요됨
- 무지 먹방 방법
    1. 1번 음식부터 먹기 시작하며, 회전판은 번호가 증가하는 순서대로 음식을 무지 앞으로 가져다 놓음
    2. 마지막 번호의 음식을 섭취한 후에는 회전판에 의해 다시 1번 음식이 무지 앞으로 옴
    3. 무지는 음식 하나를 1초 동안 섭취한 후 남은 음식은 그대로 두고, 다음 음식을 섭취함(남은 음식 중 다음으로 섭취해야 할 가장 가까운 번호의 음식)
    4. 회전판이 다음 음식을 무지 앞으로 가져오는데 걸리는 시간은 없다고 가정

- 무지가 먹방을 시작한 지 K초 후에 네트워크 장애로 인해 방송이 잠시 중단됨
- 네트워크 정상화 후 다시 방송을 이어갈 때, 몇 번 음식부터 섭취해야 하는지를 알고자 함
- 음식을 모두 먹는데 필요한 배열 food_times
- 네트워크 장애가 발생한 시간 K초(매개 변수)
- 몇 번 음식부터 다시 섭취하면 되는지 solution 함수 생성

```python
food_times = list(map(int, input().split(" ")))
k = int(input())
print(food_times, k)
```

```python
def solution(food_times, k):
    n = 0
    while n <= k:
        for i in range(len(food_times)):
            if food_times[i] != 0:
                n += 1
                food_times[i] -= 1
            else:
                continue
            if n > k:
                answer = i + 1
    if sum(food_times) == 0:
            answer = -1
        
    return answer
        
    


```

```python
from operator import itemgetter

def solution(food_times, k):
    foods = []
    n = len(food_times)
    for i in range(n):
        foods.append((food_times[i], i+1))
    
    foods.sort()

    pretime = 0
    for i, food in enumerate(foods):
        diff = food[0] - pretime
        if diff != 0:
            spend = diff * n # 소요한 시간 빼기
            if spend <= k:
                k -= spend
                pretime = food[0]
            else:
                k %= n
                sublist = sorted(foods[i:], key = itemgetter(1)) # itemgetter(1)이면 1번 인덱스를 기준으로 정렬
                return sublist[k][1]
        n -= 1

    return -1
```
