# 1. 모험가 길드
 - N명의 모험가를 대상으로 '공포도'를 측정함
 - 공포도가 X인 모험가는 반드시 X명 이상으로 구성한 모험가 그룹에 참여해야 여행을 떠날 수 있음
 - 여행을 떠날 수 있는 그룹의 최대값을 구하는 프로그램을 만드시오

```python
n = int(input())
guild = sorted(list(map(int, input().split(" "))))
guild
```

```python
result = 0 # 총 그룹의 수
count = 0 # 현재 그룹에 포함된 모험가의 수

for i in guild: # 공포도를 낮은 것부터 하나씩 확인하며
    count += 1
    if count >= i:
        result += 1
        count = 0

print(result)
```

# 2. 곱하기 혹은 더하기
- 각 자리가 숫자(0부터 9)로만 이루어진 문자열 S가 주어졌을 때, 왼쪽부터 오른쪽으로 하나씩 모든 숫자를 확인하며 숫자 사이에 'X' 혹은 '+' 연산자를 넣어 결과적으로 만들어질 수 있는 가장 큰 값을 구하시오

```python
num = list(map(int, input()))
```

```python
num
```

```python
total_sum = 0

for i in num:
    if i <= 1 or total_sum == 0:
        total_sum += i
    else:
        total_sum *= i

print(total_sum)
```

# 문자열 뒤집기
- 다솜이는 0과 1로만 이루어진 문자열 S를 가지고 있습니다. 이 문자열 S에 있는 모든 숫자를 전부 같게 만들려고 함
- 문자열 S가 주어졌을 때, 다솜이가 해야하는 횟수의 최소 횟수를 출력

```python
string = list(map(str, input()))
string
```

```python
count0 = 0
count1 = 0
for i in range(len(string)-1):
    if string[i] != string[i+1]:
        if string[i+1] == '1':
            count0 += 1
        else:
            count1 += 1

print(min(count0, count1))
```

# 만들 수 없는 금액
- N개의 동전을 이용하여 만들 수 없는 양의 정수 금액 중 최솟값을 구하는 프로그램을 작성하시오



```python
coin_count = int(input())
coin_list = list(map(int, input().split(" ")))
```

```python
coin_list.sort()
```

```python
target = 1
for i in coin_list:
    if i > target:
        break
    else:
        target += i

print(target)
```
