# Q.37 플로이드
- n개의 도시가 있고, 한 도시에서 출발하여 다른 도시에 도착하는 m개의 버스가 있다
- 각 버스는 한 번 사용할 때 필요한 비용이 있음
- 모든 도시의 쌍(A, B)에 대해서 A에서 B로 가는데 필요한 비용의 최솟값을 구하는 프로그램을 작성하시오
- n : 도시, m : 버스, (a, b, c) : (시작도시, 도착도시, 비용)

```python
INF = int(1e9) # 무한을 의미하는 값 10억 설정

# 도시, 버스 개수
n = int(input())
m = int(input())

# 2차원 리스트 모든 값을 0으로 초기화
graph = [[INF] * (n+1) for _ in range(n+1)]
```

```python
# 자기 자신에서 자기 자신으로 가는 비용은 0으로 초기화
for a in range(1, n + 1):
    for b in range(1, n + 1):
        if a == b:
            graph[a][b] = 0
```

```python
graph
```

```python
# 각 간선에 대한 정보를 입력받아, 그 값으로 초기화
for _ in range(m):
    # A에서 B로가는 비용은 C로 설정
    a, b, c = map(int, input().split(" "))
    graph[a][b] = min(graph[a][b], c)
```

```python
graph
```

```python
# 점화식에 따라 플로이드 워셜 알고리즘 수행
for k in range(1, n + 1):
    for a in range(1, n + 1):
        for b in range(1, n + 1):
            graph[a][b] = min(graph[a][b], graph[a][k] + graph[k][b])

graph
```

```python
# i -> j 못가면 0으로 저장
for i in range(n+1):
    for j in range(n+1):
        if graph[i][j] == INF:
            graph[i][j] = 0
```

```python
# 결과 출력
for a in range(1, n + 1):
    for b in range(1, n + 1):
            print(graph[a][b], end = " ")
    print()
```

# Q.38 정확한 순위
- N명의 성적을 비교하여 성적 순위를 정확히 알 수 있는 학생은 모두 몇 명인지 계산하시오
- 1번 학생 < 5번 학생
  
  3번 학생 < 4번 학생
  
  4번 학생 < 2번 학생
  
  4번 학생 < 6번 학생
  
  5번 학생 < 2번 학생
  
  5번 학생 < 4번 학생



```python
INF = 1e9

n, m = map(int, input().split())
graph = [[INF] * (n+1) for _ in range(n+1)]
```

```python
# 자기 자신 초기화
for a in range(1, n+1):
    for b in range(1, n+1):
        if a == b:
            graph[a][b] = 0
```

```python
for _ in range(m):
    a, b = map(int, input().split(" "))
    graph[a][b] = 1
```

```python
# 점화식에 따라 플로이드 워셜 알고리즘 수행
for k in range(1, n + 1):
    for a in range(1, n + 1):
        for b in range(1, n + 1):
            graph[a][b] = min(graph[a][b], graph[a][k] + graph[k][b])
```

#Q.39 화성 탐사
- 에너지를 효율적으로 사용하고자 화성 탐사 기계가 출발 지점에서 목표 지점까지 이동할 때 항상 최적의 경로를 찾도록 개발해야 함
- N X N 크기의 2차원 공간이며, 각가의 칸을 지나기 위한 비용이 존재
- 가장 왼쪽 위 칸인 [0][0] 위치에서 가장 오른쪽 아래 칸인 [N - 1][N - 1] 위치로 이동하는 최소 비용을 출력하는 프로그램을 작성하시오

```python
import heapq
import sys
INF = int(1e9) # 무한을 의미하는 값으로 10억을 설정

dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

# 전체 테스트 케이스(Test Case)만큼 반복
for tc in range(int(input())):
    # 노드의 개수를 입력받기
    n = int(input())
    # 전체 맵 정보를 입력받기
    graph = []
    for i in range(n):
        graph.append(list(map(int, input().split())))
        
    # 최단 거리 테이블을 모두 무한으로 초기화
    distance = [[INF] * n for _ in range(n)]

    x, y = 0, 0 # 시작 위치는 (0, 0)
    # 시작 노드로 가기 위한 비용은 (0, 0) 위치의 값으로 설정하여, 큐에 삽입
    q = [(graph[x][y], x, y)]
    distance[x][y] = graph[x][y]

    # 다익스트라 알고리즘 수행
    while q:
        # 가장 최단 거리가 짧은 노드에 대한 정보를 꺼내기
        dist, x, y = heapq.heappop(q)
        # 현재 노드가 이미 처리된 적이 있는 노드라면 무시
        if distance[x][y] < dist:
            continue
        # 현재 노드와 연결된 다른 인접한 노드들을 확인
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            # 맵의 범위를 벗어나는 경우 무시
            if nx < 0 or nx >= n or ny < 0 or ny >= n:
                continue
            cost = dist + graph[nx][ny]
            # 현재 노드를 거쳐서, 다른 노드로 이동하는 거리가 더 짧은 경우
            if cost < distance[nx][ny]:
                distance[nx][ny] = cost
                heapq.heappush(q, (cost, nx, ny))

    print(distance[n - 1][n - 1])    
```
