```python
import matplotlib.pyplot as plt
import math
import time
import pandas as pd
```

```python
## 1번
def log2n_graph(n):
    log2n_array = [math.log2(i) for i in range(1, n+1)]
    print("log2n:",log2n_array)
    return log2n_array

def two_squared_graph(n):
    two_sqrt_array = [2**i for i in range(1, n+1)] 
    print("2의 제곱:",two_sqrt_array)
    return two_sqrt_array

def factorial_graph(n):
    factorial_array = [math.factorial(i) for i in range(1, n+1)] 
    print("n!:",factorial_array)
    return factorial_array

def nlog2n_graph(n):
    nlogn_array = [i * math.log2(i) for i in range(1, n+1)]
    print("nlog2n:",nlogn_array)
    return nlogn_array

def n_squared_graph(n):
    n_squared_array = [i**2 for i in range(1, n+1)]
    print("n의 제곱:",n_squared_array)
    return n_squared_array
    

def total_plot(n):
    fig = plt.figure(figsize=(15,15))
    n_array = [i for i in range(1, n+1)]
    print("n의 값:",n_array)
    plt.plot(n_array, n_array, label = 'n')
    plt.plot(n_array, log2n_graph(n), label = 'log2n')
    plt.plot(n_array, two_squared_graph(n), label = '2^n')
    if n <= 4:
        plt.plot(n_array, factorial_graph(n), label = 'n!')
    plt.plot(n_array, nlog2n_graph(n), label = 'nlog2n')
    plt.plot(n_array, n_squared_graph(n), label = 'n^2')
    plt.legend()
    
    

```

```python
total_plot(5)
```

```python
## 2번
def nxn(n):
  start_time = time.time()
  for i in range(1, n+1):
    for j in range(1, n+1):
          i * j
  end_time = time.time()
  time_taken = int(end_time - start_time)
  print("걸린 시간은 {}초입니다.".format(time_taken))
  return time_taken
```

```python
## 3번
nxn10 = nxn(10)
nxn100 = nxn(100)
nxn1000 = nxn(1000)
nxn10000 = nxn(10000)
nxn100000 = nxn(100000)
```

```python
nxn_df = pd.DataFrame(data=[{"nxn10":nxn10,"nxn100":nxn100,"nxn1000":nxn1000,"nxn10000":nxn10000,"nxn100000":nxn100000}])
```
