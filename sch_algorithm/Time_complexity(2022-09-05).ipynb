```python
import matplotlib.pyplot as plt
import math
import time

## 1번
def log2n_graph(n):
    log2n_array = [math.log2(i) for i in range(1, n)]  
    return log2n_array

def two_squared_graph(n):
    two_sqrt_array = [2^i for i in range(1, n)] 
    return two_sqrt_array

def factorial_graph(n):
    factorial_array = [math.factorial(i) for i in range(1, n)] 
    return factorial_array

def nlogn_graph(n):
    nlogn_array = [n * math.log2(i) for i in range(1, n)]
    return nlogn_array

def n_squared_graph(n):
    n_squared_array = [i^2 for i in range(1, n)]
    return n_squared_array
    

def total_plot(n):
    fig = plt.figure(figsize=(15,15))
    n_array = [i for i in range(1, n)]
    plt.plot(n_array, n_array, label = 'n')
    plt.plot(n_array, log2n_graph(n), label = 'log2n')
    plt.plot(n_array, two_squared_graph(n), label = '2^n')
    if n <= 5:
        plt.plot(n_array, factorial_graph(n), label = 'n!')
    plt.plot(n_array, nlogn_graph(n), label = 'nlog2n')
    plt.plot(n_array, n_squared_graph(n), label = 'n^2')
    plt.legend()
    
    

```

```python
total_plot(5)
```

```python
## 2번
def nxn(n):
  start_time = time.time()
  for i in range(n):
    for j in range(n):
      n * j
  end_time = time.time()
  print("걸린 시간은 {}입니다.".format(end_time - start_time))
```

```python
## 3번
nxn(10)
nxn(100)
nxn(1000)
nxn(10000)
nxn(100000)
```
