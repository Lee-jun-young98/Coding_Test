# 8. 문자열 재정렬
- 알파벳 대문자와 숫자(0 ~ 9)로만 구성된 문자열이 입력으로 주어짐
- 모든 알파벳을 오름차순으로 정렬하여 이어서 출력한 뒤에, 모든 숫자를 더한 값을 이어서 출력함

```python
arr = list(input()) # 데이터 받기
arr.sort() # 오름차순으로 정렬하기
```

```python
arr
```

```python
result = []
sum = 0
for i in arr:
    if i.isalpha(): # i가 알파벳이면 결과값 행렬에 추가
        result.append(i)
    else:
        sum += int(i) # 숫자면 더하기
    
```

```python
result.sort()
```

```python
if sum != 0:
    result.append(str(sum))
```

```python
print(''.join(result))
```
