# Q.23 국영수
- N명의 이름과 국어, 영어, 수학, 점수가 주어짐
- 이 때 다음과 같은 조건으로 성적을 정렬하는 프로그램을 작성하시오
    1. 국어 점수가 감소하는 순서로
    2. 국어 점수가 같으면 영어 점수가 증가하는 순서로
    3. 국어 점수와 영어 점수가 같으면 수학 점수가 감소하는 순서로
    4. 모든 점수가 같으면 이름이 사전 순으로 증가하는 순서로(대문자는 소문자보다 작으므로 사전 순으로 앞에 오게 됨)

```python
students = []
for _ in range(int(input())):
    students.append(input().split())

students.sort(key=lambda x: (-int(x[1]), int(x[2]), -int(x[3]), x[0])) # 국어 점수가 감소하는, 영어 점수가 증가하는, 수학 점수가 감소하는, 이름이 사전순으로 적용

# 정렬된 학생 정보에서 이름만 출력
for student in students:
    print(student[0])
```
