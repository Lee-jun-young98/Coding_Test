# Q.28 고정점 찾기
- 고정점이란, 수열의 원소 중에서 값과 인덱스가 동일한 원소를 의미함
- ex) a = {-15, -4, 2, 8, 13}이 있을 때 a[2] = 2이므로, 고정점은 2가 됨
- 하나의 수열이 N개의 서로 다른 원소를 포함하고 있으며, 모든 원소가 오름 차순으로 정렬됨
- 고정점이 있으면 고정점을 출력하고 없으면 -1을 출력하는 프로그램을 작성하시오

```python
n = int(input())
arr = list(map(int, input().split(" ")))
```

```python
def find_fix_point(arr, n):
    for i in range(n):
        if arr[i] == i:
            return arr[i]
    return -1
```

```python
find_fix_point(arr, n)
```
