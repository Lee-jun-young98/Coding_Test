# 2. 곱하기 혹은 더하기
- 각 자리가 숫자(0부터 9)로만 이루어진 문자열 S가 주어졌을 때, 왼쪽부터 오른쪽으로 하나씩 모든 숫자를 확인하며 숫자 사이에 'x' 혹은 '+' 연산자를 넣어 결과적으로 만들어질 수 있는 가장 큰 값을 구하시오

```python
num = list(map(int, input()))
```

```python
num
```

```python
total_sum = 0

for i in num:
    if i <= 1 or total_sum == 0: # 1보다 작거나 0이면 *말고 더하기를 해야함 처음 시작이 0이면 +를 해야 값이 시작됨
        total_sum += i
    else:
        total_sum *= i

print(total_sum)
```
